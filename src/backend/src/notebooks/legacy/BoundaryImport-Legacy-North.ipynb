```python
from morpheus.project.types.geometry import Point
import dataclasses


@dataclasses.dataclass
class PumpingWell:
  type: str
  name: str
  geometry: Point
  layers: list[int]
  well_type: str
  sp_values: list[list[int]]
  is_excluded_from_calculation: bool

  @classmethod
  def from_geojson(cls, geojson):
    properties = geojson['properties']
    return cls(
      type='wel',
      name=str(properties['fid']),
      geometry=Point.from_dict(geojson['geometry']),
      layers=[properties['Layer'] - 1],
      well_type='puw',
      sp_values=[[properties['Abstr_max'] * -1]],
      is_excluded_from_calculation=False
    )

  def to_dict(self):
    return {
      'type': self.type,
      'name': self.name,
      'geometry': self.geometry.to_dict(),
      'layers': self.layers,
      'well_type': self.well_type,
      'sp_values': self.sp_values,
      'is_excluded_from_calculation': self.is_excluded_from_calculation
    }
```

```python
feature = geojson['features'][0]
pumping_well = PumpingWell.from_geojson(feature)
pumping_well.to_dict()
```

```python
import simplejson as json

with open(os.path.join(directory, 'pumping_well.json'), 'w') as f:
  json.dump(pumping_well.to_dict(), f, ignore_nan=True)
```

```python
pumping_wells = [PumpingWell.from_geojson(feature).to_dict() for feature in geojson['features']]
pumping_wells
```

```python
with open(os.path.join(notebook_path, 'assets', 'shapefiles', 'Mumbwa', 'pumping_wells.json'), 'w') as f:
  json.dump(pumping_wells, f, ignore_nan=True)
```
