```python
import dataclasses

from morpheus.project.types.geometry import Point


@dataclasses.dataclass
class HeadObservation:
    type: str
    name: str
    geometry: Point
    layers: list[int]
    well_type: str
    sp_values: list[list[int]]
    is_excluded_from_calculation: bool

    @classmethod
    def from_geojson(cls, geojson):
        properties = geojson['properties']
        name = str(properties['fid_1'])
        if (geojson['geometry']) is None:
            print(f'Geometry None for {name}')
            return

        return cls(
            type='hob',
            name=name,
            geometry=Point.from_dict(geojson['geometry']),
            layers=[properties['Layer'] - 1],
            well_type='puw',
            sp_values=[[properties['Obs_head']]],
            is_excluded_from_calculation=False,
        )

    def to_dict(self):
        return {
            'type': self.type,
            'name': self.name,
            'geometry': self.geometry.to_dict(),
            'layers': self.layers,
            'date_times': ['2000-01-01'],
            'sp_values': self.sp_values,
            'is_excluded_from_calculation': self.is_excluded_from_calculation,
        }
```

```python
feature = geojson['features'][0]
head_observation = HeadObservation.from_geojson(feature)
head_observation.to_dict()
```

```python
head_observations = [HeadObservation.from_geojson(feature).to_dict() for feature in geojson['features'] if feature['geometry'] is not None]
head_observations
```

```python
import simplejson as json

with open(os.path.join(directory, filename + '.json'), 'w') as f:
    json.dump(head_observations, f, ignore_nan=True)
```
